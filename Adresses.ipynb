```python
import pandas as pd
import sqlite3
import numpy as np
from db import db

# We open the db
myDB = db.DB("./mydb.db")

# A helper function to convert numpy type to string 
def to_str(var):
    return str(list(np.reshape(np.asarray(var), (1, np.size(var)))[0]))[1:-1] 
```

```python
# We open our CSV
data = pd.read_csv('Climat-2021-2051-8.5-ete-hiver-raw.csv');
```

```python
# Let's see what's inside
data.head()
```

```python
# We add a col to our table with the matching address form our db

def get_address(point):
    try:
        res = myDB.get("SELECT address FROM address WHERE point = "+to_str(point)+" LIMIT 1")
        return res[0][0]
    except:
        return "not found"
        

data['address']  = data.apply(lambda row: get_address(row[0]), axis=1)
    
data
    
```

```python
# We save our file 

data.to_csv("with_address.csv", encoding='utf-8', index=False)
```

```python

# We add a col to our table with the matching postalcode form our db

def get_postal(point):
    try:
        res = myDB.get("SELECT codepostal FROM address WHERE point = "+to_str(point)+" LIMIT 1")
        return res[0][0]
    except:
        return "not found"
        
data['postalcode'] =  data.apply(lambda row: get_postal(row[0]), axis=1)

data
```

```python
# We save our file 
data.to_csv("with_address_postal_code.csv", encoding='utf-8', index=False)
```
